# Tournament of different algorithm on the Tank Game


```python
# imports

import numpy as np

from MCTS.elements import Board
from MCTS.methods import FlatMC, UCT, UCB
```

```python
def game(display=False):
    b = Board(mode="2p")
    b.reset()
    iterate = 1
    while not b.terminal():
        flat = FlatMC(b, "p1", 100)
        m = flat.algo()
        b.play(m)
        # b.print()
        #        if iterate % 10 == 0:
        if not b.terminal():
            uct = UCT(b, "p2", 100)
            m = uct.algo()
            b.play(m)
            # b.print()
        iterate += 1
    print(f"Game done in {iterate} steps")
    print("Player 1:", b.state["player"].kills, "kills")
    print("Player 2:", b.state["enemy"].kills, "kills")
    if display:
        b.print()
    return b.score()


s = {"p1": [], "p2": []}
iterations = 50

i = 0
display = i % 20 == 0 or i == iterations - 1

for i in range(iterations):
    score = game()
    s["p1"].append(round(score["p1"]))
    s["p2"].append(round(score["p2"]))

    print("score:", s["p1"][i], s["p2"][i])

npscore = np.array([s["p1"], s["p2"]])

print("Average score:", npscore.mean(axis=1))
```
