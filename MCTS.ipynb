# Tournament of different algorithm on the Tank Game


```python
# imports

from MCTS.elements import Board
from MCTS.methods import FlatMC, UCT, UCB


def game():
    b = Board(mode="2p")
    b.reset()
    iterate = 1
    while not b.terminal():
        flat = FlatMC(b, "p1", 10)
        m = flat.algo()
        b.play(m)
        # b.print()
        #        if iterate % 10 == 0:
        if not b.terminal():
            uct = UCT(b, "p2", 10)
            m = uct.algo()
            b.play(m)
            # b.print()
        iterate += 1
    print(f"Game done in {iterate} steps")
    b.print()
    return b.score()


s = {"p1": 0, "p2": 0}
for i in range(10):
    score = game()
    s["p1"] += score["p1"]
    s["p2"] += score["p2"]

    print("score:", s)
```
